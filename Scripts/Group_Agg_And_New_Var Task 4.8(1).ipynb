```python
# Question 1. + 2. - finding the aggregated mean of the “order_number” column grouped by “department_id” for the ENTIRE dataframe
# Importing 'ords_prods_merge' for creating/adding the two new columns from exercise 4.7
df_ords_prods_merge = pd.read_pickle(os.path.join(path, 'Data','Prepared Data', 'ords_prods_merge.pkl'))
```

```python
df_ords_prods_merge.shape
```

```python
# Frequency check of 'order_dow'
df_ords_prods_merge["order_dow"].value_counts(dropna = False)
```

```python
# Dropping missing values
df_ords_prods_merge_cleaned = df_ords_prods_merge.dropna(subset=["order_dow"])
```

```python
df_ords_prods_merge_cleaned["order_dow"].value_counts(dropna = False)
```

```python
# Assigning the string values
result = []

for value in df_ords_prods_merge_cleaned["order_dow"]:
  if value in (0, 1):
    result.append("Two busiest days")
  elif value in (3, 4):
    result.append("Two slowest days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Creating the new column and merging it with the result values
df_ords_prods_merge_cleaned["Busiest days"] = result
```

```python
# Frequency check
df_ords_prods_merge_cleaned["Busiest days"].value_counts(dropna = False)
```

```python
# Accuracy check
df_ords_prods_merge_cleaned["Busiest days"].describe()
```

```python
# New column 'busiest_period_of_day'
# Frequency check of 'order_hour_of_day'
df_ords_prods_merge_cleaned["order_hour_of_day"].value_counts(dropna = False)
```

```python
# Assigning the new string values
result2 = []

for value in df_ords_prods_merge_cleaned["order_hour_of_day"]:
  if value in (10, 11, 14, 15, 13, 12, 16, 9):
    result2.append("Most orders")
  elif value in (3, 4, 2, 5, 1, 0, 6, 23):
    result2.append("Fewest orders")
  else:
    result2.append("Average orders")
```

```python
result2
```

```python
# Creating the new column within the dataframe and merging it with the result2 values
df_ords_prods_merge_cleaned["busiest_period_of_day"] = result2
```

```python
# Frequency check
df_ords_prods_merge_cleaned["busiest_period_of_day"].value_counts(dropna = False)
```

```python
# Exporting the entire dataframe
df_ords_prods_merge_cleaned.to_pickle(os.path.join(path, 'Data', 'Prepared Data', 'Pr_ords_prods_merge_new_var.pkl'))
```
